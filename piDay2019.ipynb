### Ramanujan- und Ramanujanâ€“Sato-Reihe

$$\frac{1}{\pi}=\frac{2\sqrt{2}}{9801}\sum_{k=0}^\infty \frac{(4k)!(1103+26390k)}{(k!)^4 396^{4k}}=\frac{2\sqrt{2}}{99^2}\sum_{k=0}^\infty \frac{(4k)!}{(k!)^4} \frac{26390k+1103}{396^{4k}}$$

```python
# Ramanujan-Reihe
import math
from decimal import getcontext, Decimal
getcontext().prec=6

x=0

for k in range(8):
    x += Decimal(math.factorial(4*k)*(1103+26390*k)) / Decimal((math.factorial(k)**4*396**(4*k)))
    print(k,":",9801/(2*Decimal(2).sqrt()*x))
```

```python
# AGM
# Arithmetisches und geometrisches Mittel

from decimal import getcontext, Decimal
getcontext().prec=50

a = 1
b = 1/Decimal(2).sqrt()
s = 1/Decimal(4)

for n in range(7):
    #print(n, "a,b:",a, b)
    A = (a+b)/2
    B = Decimal(a*b).sqrt()
    S = s - 2**n * (a-A)**2
    #print(n, "A,B:",A, B)
    print(n,":",A**2/s)
    
    a = A
    b = B
    s = S
    
```
