```python
def fib(n):
    if n==1 or n==2:
       r=1
    else:
       r=fib(n-1)+fib(n-2) 
    return r

def fib_td(n):
    f1=1; f2=1;f=1
    for i in range(3, n+1):
        f=f1+f2
        #print (i,f2,f1,f)
        f2=f1; f1=f
    return f
```
