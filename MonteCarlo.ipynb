```python
import random
import time
#szaehler = input("Mit wievielen Zufallszahlen soll versucht werden, pi zu finden?: ")
counter_in = 0
zaehler=100000000 # int(szaehler)
t0=time.time()
for i in range(zaehler):
    x = random.random()
    y = random.random()
    
    if x*x + y*y < 1:
        counter_in += 1
t1=time.time()-t0
        
pi = float(4*counter_in)/zaehler
 
print (pi, t1)
```
