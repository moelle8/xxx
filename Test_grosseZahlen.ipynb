```python
#!/usr/bin/python2
# -*- coding: utf-8 -*-

# berechnet die p-te Potenz von start durch Multiplikation 
# und rechnet durch Division wieder zurueck, Ergebnis muss
# dann gleich start sein

p = 13
start = 8648234678954023643289846329847304534823

sum = start

for i in range(p):
    sum = sum * start

print(sum)
    
for i in range(p):
    sum = sum // start

print(sum)
print(start)
```
