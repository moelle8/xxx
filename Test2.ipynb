```python
def primfaktorzerlegung(zahl):
    rest = zahl
    primfaktoren = []
    while rest != 1:
        a = int(rest ** 0.5)
        for j in range(2, a + 1):
            if rest % j == 0:
                primfaktoren += [j]
                rest = int(rest / j)
                break
        else:
            primfaktoren += [rest]
            break
    return primfaktoren
 
 
from collections import Counter
 
a = primfaktorzerlegung(2)
for i in range(3, 99):
    b = primfaktorzerlegung(i)
    c = Counter(a) | Counter(b)
    a = list(c.elements())
 
produkt = 1
for i in a:
    produkt *= i
print(produkt)
```
