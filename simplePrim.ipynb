```python
def isPrim(n):
    # Test 2,3
    r=0
    n2=int(n**.5+1.1)
    for i in range(3,n2,2):
        if n % i == 0: r=i; break
    return r

def isPrim2(n):
    # Test 2,3
    r=0
    n2=int(n**.5+1.1)
    i=5; di=2
    while i<n2:
        if n % i == 0: r=i; break
        i=i+di; di=6-di
    return r
```

```python
#%timeit isPrim(2**31-1)
import time
t1=time.time()
isPrim(2**61-1) # 106.862 s
t1=time.time()-t1

print(t1)
```
